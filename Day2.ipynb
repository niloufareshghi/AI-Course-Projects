```python
n = int(input())
x = 0
for i in range(n):
  s = input()
  if s == "X++":
    x = x + 1
  if s == "X--":
    x = x - 1
  if s == "--X":
    x = x - 1
  if s == "++X":
    x = x + 1
print(x)
```

```python
x = '1'
for k in range(5):
  l = [item for item in input().split()]
  if x in l:
    i = k + 1
    j = l.index(x) + 1
out = abs(3 - i) + abs(3 - j)
print(out)
```

```python
one = input().lower()
two = input().lower()
if one == two:
  print(0)
else:
  if one < two:
    print(-1)
  if two <= one:
    print(1)
```

```python
t = int(input())
for i in range(t):
  n = int(input())
  count = 0
  d = len(str(n))
  count += (d-1)*9
  l = [int(str(j)*d) for j in range(1, 10)]
  for item in l:
    if item <= n:
      count +=1
  print(count)
```
